## 定义超参数

```python
DATSET = "../data/women-clothes"

START_TRAIN_IMG_SIZE = 8
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
LR = 1e-3
BATCH_SIZES = [256,128,64,32,16,8]
CHANNELS_IMG = 3
#以下三个参数设置为256，为了使用较少内存，有条件的也可设置更大一些，如512
Z_DIM = 256
W_DIM = 256
IN_CHANNELS = 256
#这个参数为wgan-gp损失中参数
LAMBDA_GP = 10
PROGRESSIVE_EPOCHS = [30] * len(BATCH_SIZES)
```

## 对数据进行预处理
将图像调整到我们想要的分辨率，将其转换为张量，然后进行一些增强，最后将其归一化为-1到1之间的所有像素。使用DataLoader创建小批量大小，DataLoader获取数据集，并通过混洗数据来创建批量大小。

```python
def get_loader(image_size):
    trainsform = transforms.Compose(
        [transforms.Resize((image_size, image_size)),
         transforms.ToTensor(),
         transforms.RandomHorizontalFlip(p=0.5),
         transforms.Normalize(
            [0.5 for _ in range(CHANNELS_IMG)],
            [0.5 for _ in range(CHANNELS_IMG)],
         )
        ]
    )
    batch_size = BATCH_SIZES[int(log2(image_size/4))]
    dataset = dset.ImageFolder(root=DATSET, transform=trainsform)
    loader =torch.utils.data.DataLoader(dataset,batch_size=batch_size,shuffle=True)
    return loader, dataset
```

## StyleGAN的训练过程



定义一个factors变量，该变量包含将与IN_CHANNELS相乘的数字，以获得在每个图像分辨率中想要的通道数。

```python
# Noice apping network
# AdaIN
# Progressice growing
factors = [1,1,1,1/2,1/4,1/8,1/16,1/32]
```

## 构建噪声映射网络

噪声映射网络读取Z，并将其穿过八个完全连接的层，这些层通过一些激活而分离。不要忘记像作者在ProGAN中所做的那样均衡学习率（ProGAN和StyleGan由同一研究人员撰写）。



让我们首先构建一个名为WSLinear（加权缩放线性）的类，它将继承自nn.Module。
在init部分，输入In_features和out_channels。创建一个线性层，然后我们定义一个比例，该比例将等于2除以in_features的平方根，将当前列层的偏移复制到一个变量中，因为我们不希望线性层的偏移被缩放，然后我们将其移除，最后，初始化线性层。
在前向部分，输入x，对其进行缩放，并添加偏差。仿射变换的输出给ADaIN做后续处理。

```python
class WSLinear(nn.Module):
    def __init__(
        self, in_features, out_features
    ):
        super(WSLinear,self).__init__()
        self.linear = nn.Linear(in_features, out_features)
        self.scale  = (2/in_features) ** 0.5
        self.bias   = self.linear.bias
        self.linear.bias = None
        
        #对权重和偏置进行初始化
        nn.init.normal_(self.linear.weight)
        nn.init.zeros_(self.bias)

    def forward(self,x):
        #对x进行缩放
        return self.linear(x * self.scale) + self.bias
```

在stylegan中,初始特征向量z会先经过pixelnorm再流向mapping层转换成线性无关的中间特征向量.
pixelnorm公式如下：

选择PixelNorm原因:是因为它没有需要训练的参数,减轻了开销;
并且normalizes了特征向量的每个元素到单位长度附近,阻止了signal magnitudes导致的在训练过程中逐步失控的风险

```python
class PixenNorm(nn.Module):
    def __init__(self):
        super(PixenNorm, self).__init__()
        self.epsilon = 1e-8
    def forward(self,x ):
        return x / torch.sqrt(torch.mean(x**2, dim=1, keepdim=True)+  self.epsilon)
```

## 构建映射网络
由8个全连接层构成


```python
class MappingNetwork(nn.Module):
    def __init__(self, Z_DIM, w_dim):
        super().__init__()
        self.mapping = nn.Sequential(
            PixenNorm(),
            WSLinear(Z_DIM, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
            nn.ReLU(),
            WSLinear(w_dim, w_dim),
        )
    
    def forward(self,x):
        return self.mapping(x)
```

## 实现AdaIN类
Adaptive Instance Normalization (AdaIN),
在初始化部分，输入通道w_dim，并初始化instance_norm，它将是实例规范化部分，
初始化style_scale和style_bias，它们将是WSLinear的自适应部分，WSLinear将噪声映射网络w映射到通道中。
在前向传递中，输入x，对其应用实例规范化，并返回style_sclate*x+style_bias

## 实现AdaIN功能


```python
class AdaIN(nn.Module):
    def __init__(self, channels, w_dim):
        super().__init__()
        #对x实例正则化
        self.instance_norm = nn.InstanceNorm2d(channels)
        #对w进行仿射变换，生成ys，yb
        self.style_scale   = WSLinear(w_dim, channels)
        self.style_bias    = WSLinear(w_dim, channels)

    def forward(self,x,w):
        x = self.instance_norm(x)
        style_scale = self.style_scale(w).unsqueeze(2).unsqueeze(3)
        style_bias  = self.style_bias(w).unsqueeze(2).unsqueeze(3)
        return style_scale * x + style_bias
```

## 创建类InjectNoise
实现图中X+B的功能，创建类InjectNoise，将噪声注入生成器  
在初始化部分，输入通道数，从随机正态分布初始化权重，并使用nn.Parameter来优化这些权重。
在前向部分，我们发送一个图像x，并在返回时添加随机噪声

```python
class injectNoise(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.weight = nn.Parameter(torch.zeros(1,channels,1,1))

    def forward(self, x):
        noise = torch.randn((x.shape[0], 1, x.shape[2], x.shape[3]), device = x.device)
        return x + self.weight + noise
```

## 构建生成模块
生成模块由conv->AdaIN->conv->AdaIN构成
在生成器体系结构中，我们有一些重复的模式，所以需先为它创建一个类，以使整个代码尽可能简单明了，将从nn.Module继承的类命名为GenBlock。  
在初始化部分中，输入In_channels、out_channelss和w_dim，然后通过将In_channel映射到out_channel的WSConv2d初始化conv1，通过将out_channels映射到out_channels的WSConf2d初始化conv2，通过以0.2的斜率定义激活函数leaky ReLU，通过InjectNoise初始化inject_noise1、inject_nouse2，通过AdaIN初始化adain1和adain2。
在前向部分，输入x，并将其传递给conv1，然后用leakay注入_noise1，然后用adain1对其进行归一化，再次将其传递到conv2，然后用Leakay注入_noise2，并用adain2对其进行规范化。最后，返回x。



```python
class GenBlock(nn.Module):
    def __init__(self, in_channel, out_channel, w_dim):
        super(GenBlock, self).__init__()
        self.conv1 = WSConv2d(in_channel, out_channel)
        self.conv2 = WSConv2d(out_channel, out_channel)
        self.leaky = nn.LeakyReLU(0.2, inplace=True)
        self.inject_noise1 = injectNoise(out_channel)
        self.inject_noise2 = injectNoise(out_channel)
        self.adain1 = AdaIN(out_channel, w_dim)
        self.adain2 = AdaIN(out_channel, w_dim)
    def forward(self, x,w):
        x = self.adain1(self.leaky(self.inject_noise1(self.conv1(x))), w)
        x = self.adain2(self.leaky(self.inject_noise2(self.conv2(x))), w)
        return x
```

## 构建生成器


```python
class Generator(nn.Module):
    def __init__(self, Z_DIM, w_dim, in_channels, img_channels=3):
        super().__init__()
        #生成常数项输入，形状为4x4
        self.starting_cte = nn.Parameter(torch.ones(1, in_channels, 4,4))
        self.map = MappingNetwork(Z_DIM, w_dim)
        self.initial_adain1 = AdaIN(in_channels, w_dim)
        self.initial_adain2 = AdaIN(in_channels, w_dim)
        #对应图中x+B
        self.initial_noise1 = injectNoise(in_channels)
        self.initial_noise2 = injectNoise(in_channels)
        self.initial_conv   = nn.Conv2d(in_channels, in_channels, kernel_size=3, stride=1, padding=1)
        self.leaky          = nn.LeakyReLU(0.2, inplace=True)

        self.initial_rgb    = WSConv2d(
            in_channels, img_channels, kernel_size = 1, stride=1, padding=0
        )
        self.prog_blocks, self.rgb_layers = (
            nn.ModuleList([]),
            nn.ModuleList([self.initial_rgb])
        )

        for i in range(len(factors)-1):
            conv_in_c  = int(in_channels * factors[i])
            conv_out_c = int(in_channels * factors[i+1])
            self.prog_blocks.append(GenBlock(conv_in_c, conv_out_c, w_dim))
            self.rgb_layers.append(WSConv2d(conv_out_c, img_channels, kernel_size = 1, stride=1, padding=0))
        
    def fade_in(self, alpha, upscaled, generated):
        return torch.tanh(alpha * generated + (1-alpha ) * upscaled)

    def forward(self, noise, alpha, steps):
        w = self.map(noise)
        #第一个adain
        x = self.initial_adain1(self.initial_noise1(self.starting_cte),w)
        x = self.initial_conv(x)
        #其他adain
        out = self.initial_adain2(self.leaky(self.initial_noise2(x)), w)

        if steps == 0:
            return self.initial_rgb(x)
        
        for step in range(steps):
            #进行上采样，每次扩充2倍
            upscaled = F.interpolate(out, scale_factor=2, mode = 'bilinear')
            out      = self.prog_blocks[step](upscaled,w)

        final_upscaled = self.rgb_layers[steps-1](upscaled)
        final_out      = self.rgb_layers[steps](out)

        return self.fade_in(alpha, final_upscaled, final_out)
```

在初始化部分中，让我们通过常数4 x 4（原论文为x512通道，考虑到减少资源使用，也设置为256）来初始化“starting_constant”张量，该张量通过生成器的迭代放置，通过“MappingNetwork”进行映射，通过AdaIN进行initial_ain1、initial_ain2、通过InjectNoise进行initial_noise1、initiat_noise2、通过将in_channels映射到自身的conv层进行初始化，leaky ReLU以0.2的斜率泄漏，WSConv2d的initial_rgb将in_channels映射到img_channels wi=（对于RGB为3），ModuleList（）的prog_blocks将包含所有渐进块（我们通过乘以in_channel来指示卷积输入/输出通道，在我们的情况下，in_channels为512，在因子的情况下为256），以及ModuleList的rgb_blocks（）将包含所有RGB块。
为了在新的层中注入（ProGAN的一个原始组件），我们添加了fade_in部分，我们发送alpha、缩放和生成它，然后返回。
tanh(alphagenerated+(1-alpha)upscale)
使用tanh的原因是它将是输出（生成的图像），并且希望像素在1到-1之间。 在前向部分，我们发送噪声（Z_dim），即在训练期间将缓慢衰减的阿尔法值（阿尔法在0和1之间），以及我们正在处理的当前分辨率的步长，我们将x传递到映射中以获得中间噪声向量W，我们将starting_constant传递到initial_noise1，应用它和W initial_ain1，然后我们将其传递到initial_cv中，再次为其添加initial_noise2，并将leak作为激活函数，然后应用它和W initial_ain2。然后我们检查步长是否为0，如果是，那么我们所要做的就是通过初始RGB运行它，我们已经完成了，否则，我们在步长上循环，在每个循环中，我们放大（放大），并通过与该分辨率相对应的渐进块运行（out）。最后，我们返回fade_In，它在将alpha、final_out和final_upscaled映射到RGB后进行处理。  

## 构建卷积层

```python
class WSConv2d(nn.Module):
    def __init__(
        self, in_channels, out_channels, kernel_size=3, stride=1, padding=1
    ):
        super(WSConv2d, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)
        self.scale = (2 / (in_channels * (kernel_size ** 2))) ** 0.5
        self.bias = self.conv.bias
        self.conv.bias = None

        # 初始化卷积层
        nn.init.normal_(self.conv.weight)
        nn.init.zeros_(self.bias)

    def forward(self, x):
        #对卷积的输入进行缩放
        return self.conv(x * self.scale) + self.bias.view(1, self.bias.shape[0], 1, 1)
```

## 定义卷积模块

```python
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(ConvBlock, self).__init__()
        self.conv1 = WSConv2d(in_channels, out_channels)
        self.conv2 = WSConv2d(out_channels, out_channels)
        self.leaky = nn.LeakyReLU(0.2)

    def forward(self, x):
        #卷积模块有两个卷积层构成，使用leaky激活函数
        x = self.leaky(self.conv1(x))
        x = self.leaky(self.conv2(x))
        return x
```

## 构建判别器
在下面的代码片段中，构建与ProGAN中相同的类Discriminattowich

```python
class Discriminator(nn.Module):
    def __init__(self, in_channels, img_channels=3):
        super(Discriminator, self).__init__()
        self.prog_blocks, self.rgb_layers = nn.ModuleList([]), nn.ModuleList([])
        self.leaky = nn.LeakyReLU(0.2)

        # 从factors中逆推，因为判别器应该是生成器的镜像。因此，
        #附加的第一个 prog_block 和 rgb 层将适用于输入大小 1024x1024，然后是 512->256-> 等
        for i in range(len(factors) - 1, 0, -1):
            conv_in = int(in_channels * factors[i])
            conv_out = int(in_channels * factors[i - 1])
            self.prog_blocks.append(ConvBlock(conv_in, conv_out))
            self.rgb_layers.append(
                WSConv2d(img_channels, conv_in, kernel_size=1, stride=1, padding=0)
            )

        # “initial_rgb”是4x4 输入大小的RGB层
        #这样做是为了“镜像”生成器initial_rgb

        self.initial_rgb = WSConv2d(
            img_channels, in_channels, kernel_size=1, stride=1, padding=0
        )
        self.rgb_layers.append(self.initial_rgb)
        self.avg_pool = nn.AvgPool2d(
            kernel_size=2, stride=2
        )  #使用平均池化实现下采样

        # 输入形状是4x4
        self.final_block = nn.Sequential(
            # 把1加到 in_channels， 因需要与 MiniBatch std 拼接
            WSConv2d(in_channels + 1, in_channels, kernel_size=3, padding=1),
            nn.LeakyReLU(0.2),
            WSConv2d(in_channels, in_channels, kernel_size=4, padding=0, stride=1),
            nn.LeakyReLU(0.2),
            WSConv2d(
                in_channels, 1, kernel_size=1, padding=0, stride=1
            ),  # 使用卷积替换全连接层
        )

    def fade_in(self, alpha, downscaled, out):
        """ 使用avg pooling和CNN的输出实现缩放"""
        # alpha是[0, 1]范围内的标量，并且upscale.shape == generated.shape
        return alpha * out + (1 - alpha) * downscaled

    def minibatch_std(self, x):
        batch_statistics = (
            torch.std(x, dim=0).mean().repeat(x.shape[0], 1, x.shape[2], x.shape[3])
        )
        # 对每个示例（跨所有通道和像素）采用标准差，然后对单个通道重复它，并将其与图像连接起来。
        #通过这种方式，判别器将获得有关批次/图像变化的信息
        return torch.cat([x, batch_statistics], dim=1)

    def forward(self, x, alpha, steps):
        # 从 prog_blocks 列表中开始，如果steps=1，那么我们应该开始
        # 在倒数第二个，因为 input_size 将为8x8。如果steps==0我们就使用4x4块
        cur_step = len(self.prog_blocks) - steps

        # 从 RGB 转换作为初始步骤，
        # 这将取决于图像大小（每个图像都位于 RGB 层上）
        out = self.leaky(self.rgb_layers[cur_step](x))

        if steps == 0:  # i.e, image is 4x4
            out = self.minibatch_std(out)
            return self.final_block(out).view(out.shape[0], -1)

        # 因为 prog_blocks 可能会改变通道，为了缩小规模，可以使用rgb_layer
        # 来自先前/较小的尺寸，本例子与索引+1相关
        downscaled = self.leaky(self.rgb_layers[cur_step + 1](self.avg_pool(x)))
        out = self.avg_pool(self.prog_blocks[cur_step](out))

        # fade_in在采样和输入之间完成，这与生成器相反
        out = self.fade_in(alpha, downscaled, out)

        for step in range(cur_step + 1, len(self.prog_blocks)):
            out = self.prog_blocks[step](out)
            out = self.avg_pool(out)

        out = self.minibatch_std(out)
        return self.final_block(out).view(out.shape[0], -1)
```

## 通过生成器生成样本
生成器的输入包括：
- 噪音B
- 权重向量w
- 常数

在下面的代码片段中，构建generate_examples函数，该函数使用generate-gen、识别当前分辨率的步骤数以及n=100的数字。此函数的目标是生成n个伪图像并将其保存为结果。

```python
def generate_examples(gen, steps, n=100):

    gen.eval()
    alpha = 1.0
    for i in range(n):
        with torch.no_grad():
            noise = torch.randn(1, Z_DIM).to(DEVICE)
            img = gen(noise, alpha, steps)
            if not os.path.exists(f'saved_examples/step{steps}'):
                os.makedirs(f'saved_examples/step{steps}')
            save_image(img*0.5+0.5, f"saved_examples/step{steps}/img_{i}.png")
    gen.train()
```

## 求插值样本的惩罚梯度
在下面的代码片段中，采用了WGAN-GP损失的gradient_penalty函数。

```python
def gradient_penalty(critic, real, fake, alpha, train_step, device="cpu"):
    BATCH_SIZE, C, H, W = real.shape
    #从均匀分布中获取参数beta
    beta = torch.rand((BATCH_SIZE, 1, 1, 1)).repeat(1, C, H, W).to(device)
    #生成基于真实样本与生成样本的插值样本或图像
    interpolated_images = real * beta + fake.detach() * (1 - beta)
    interpolated_images.requires_grad_(True)

    # 插值样本代入判别器，获取判别分数
    mixed_scores = critic(interpolated_images, alpha, train_step)
 
    # 求关于判别分数的导数
    gradient = torch.autograd.grad(
        inputs=interpolated_images,
        outputs=mixed_scores,
        grad_outputs=torch.ones_like(mixed_scores),
        create_graph=True,
        retain_graph=True,
    )[0]
    gradient = gradient.view(gradient.shape[0], -1)
    gradient_norm = gradient.norm(2, dim=1)
    gradient_penalty = torch.mean((gradient_norm - 1) ** 2)
    return gradient_penalty
```

## 训练模型
定义训练函数  
对于train函数，我们为生成器和critic发送critic（它是判别器）、gen（生成器）、loader、dataset、step、alpha和优化器。
我们首先循环使用DataLoader创建的所有小批量大小，我们只拍摄图像，因为我们不需要标签。
然后，当我们想要最大化E（E(critic(real)) - E(critic(fake))）时，我们为鉴别器\评论家设置训练。这个等式意味着评论家能在多大程度上区分真实和虚假的图像。
之后，当我们想要最大化E(critic(fake))时，我们为生成器设置训练。
最后，我们更新fade_in的循环和alpha值，并确保它在0和1之间，然后返回它。

```python
def train_fn(
    critic,
    gen,
    loader,
    dataset,
    step,
    alpha,
    opt_critic,
    opt_gen
):
    loop = tqdm(loader, leave=True)

    for batch_idx, (real, _) in enumerate(loop):
        real = real.to(DEVICE)
        cur_batch_size = real.shape[0]
        #从标准高斯分布中采样
        noise = torch.randn(cur_batch_size, Z_DIM).to(DEVICE)
        fake  = gen(noise, alpha, step)
        critic_real = critic(real, alpha, step)
        critic_fake = critic(fake.detach(), alpha, step)
        gp = gradient_penalty(critic, real, fake, alpha, step, DEVICE)
        #创建判别器损失函数
        loss_critic = (
            -(torch.mean(critic_real) - torch.mean(critic_fake))
            + LAMBDA_GP * gp
            + (0.001) * torch.mean(critic_real ** 2)
        )

        critic.zero_grad()
        loss_critic.backward()
        opt_critic.step()

        gen_fake = critic(fake, alpha, step)
        #生成器损失函数
        loss_gen = -torch.mean(gen_fake)

        gen.zero_grad()
        loss_gen.backward()
        opt_gen.step()

        alpha += cur_batch_size / (
            PROGRESSIVE_EPOCHS[step] * 0.5 * len(dataset)
        )
        alpha = min(alpha,1)


        loop.set_postfix(
            gp = gp.item(),
            loss_critic = loss_critic.item()
        )
    return alpha
```

现在，一切准备就绪，最后把它们放在一起训练StyleGAN。  
首先初始化生成器、鉴别器/评论家和优化器，然后将生成器和评论家转换为训练模式，在PROGRESSIVE_EPOCHS上循环，
在每个循环中，调用训练函数的历元次数，然后生成一些假图像并保存它们，因此，使用generate_examples函数，最后，循环到下一个图像分辨率。

```python
gen = Generator(
    Z_DIM, W_DIM, IN_CHANNELS, CHANNELS_IMG
).to(DEVICE)
critic = Discriminator(IN_CHANNELS, CHANNELS_IMG).to(DEVICE)
opt_gen = optim.Adam([{'params': [param for name, param in gen.named_parameters() if 'map' not in name]},
                     {'params': gen.map.parameters(), 'lr': 1e-5}], lr=LR, betas =(0.0, 0.99))
opt_critic = optim.Adam(
    critic.parameters(), lr= LR, betas =(0.0, 0.99)
)

gen.train()
critic.train()
step = int(log2(START_TRAIN_IMG_SIZE / 4))
for num_epochs in PROGRESSIVE_EPOCHS[step:]:
    alpha = 1e-7
    loader, dataset = get_loader(4*2**step)
    print('Curent image size: '+str(4*2**step))

    for epoch in range(num_epochs):
        print(f'Epoch [{epoch + 1}/ {num_epochs}')
        alpha = train_fn(
            critic, gen, loader, dataset, step, alpha, opt_critic, opt_gen
        )
    generate_examples(gen, step)
    step +=1
```
