### 3.3 如何访问列表元素
如何访问列表元素，理解了列表中的元素及其对应下标（或索引），就迎刃而解了。下图就揭示了元素与对应的索引之间的一一对应关系。


#### 3.3.1 获取一个元素

```python
#创建列表a
a=[2,3,4,2,6,7]
#获取列表第1个元素（一般指从左到右，除非有特别说明）
print(a[0])
#获取列表第2个元素
print(a[1])
#获取列表最后一个元素
print(a[5])
```

列表索引，可以从左到右，也可以从右到左，如下图所示


获取最后一个元素的几种方法

```python
#（1）直接利用最后一个元素的索引
print(a[5])
#（2）如果列表比较长，最后一个索引不是很明显，可以通过计算的方式得到最后一个元素的索引：len(a)-1
print(a[len(a)-1])
#（3）更简单直观的方式是用-1索引,这种方法经常使用
print(a[-1])
```

#### 3.3.2 获取连续多个元素
对列表切片的格式为：  
<mark style=background-color:yellow>list_name[start:end[:step]] </mark>  
start表示起始索引（缺省为0），end表示结束索引，step表示步长，缺省为1

```python
#从第1个元素开始，连续取3个元素
print(a[:3])   #注意最后这个索引是3，而不是2，这里范围是[0,3)
#不是从第3个元素开始，连续取3个元素
print(a[2:5])
#不是从第3个元素开始以后所有元素
print(a[2:])
```

#### 3.3.3 遍历列表
列表是可迭代的对象（有关可迭代的对象后续将介绍）我们可以利用循环，逐一取出每个元素，然后对每个元素进行后续处理，这对使用和维护列表非常方便。

```python
#创建列表
lst1=["北京","上海","广州","深圳"]
#遍历列表，并用竖线分割
for item in lst1:
    print(item,end='|')
print()    
print("====遍历索引及对应元素====")
    
#遍历列表，同时显示索引及对应元素
for index,item in enumerate(lst1):
    print(index,item)
```

#### 3.3.4 访问列表经常出现的一个问题
我们在访问列表经常遇到list index out of range这个问题，出现这个问题，主要是访问的索引超出列表范围。如下示例，如何避免类似问题？可以借助len函数(这个函数非常有用，我们经常能看到它的身影)，先介绍列表长度，然后访问。

```python
lst11=[1,2,3,4]
#print(lst11[4])  #报错：IndexError: list index out of range
#创建一个空列表
lst12=[]  
print(lst12[0])  #报错:IndexError: list index out of range
```

```python
#借助len函数，可有效避免此类问题
lst1=["北京","上海","广州","深圳"]

for i in range(0,len(lst1)):
    print(i,lst1[i])
```

### 3.4 对列表进行增、删、改
列表是序列，而且是可以修改的序列，列表对象自身带有很多操作函数，使用这些函数就可对列表进行增加、删除、修改等操作。

#### 3.4.1 添加新元素到列表
往列表中添加新元素方法很多，如以追加的方式加入、以插入的方式加入、还可以拼接两个列表的加入等等。对应的列表函数有append、insert、extend等，具体请参考书中表3-1。

```python
#创建一个列表
lst2=['Python','Java','C++']
#添加keras
lst2.append('keras')
print(lst2)
#在索引为3的位置，插入一个元素'Pytorch'
lst2.insert(3,'Pytorch')
print(lst2)
#创建另一个列表
lst3=['TensorFlow','Caffe2']
#把列表lst3的所有元素添加到列表lst2的末尾
lst2.extend(lst3)
print(lst2)
```

#### 3.4.2 从列表中删除元素
删除列表中的元素，你可以根据位置或值来删除列表的元素。删除常用命令如下：


```python
#创建一个列表，先创建一个空列表，然后使用for循环往列表中添加元素
#创建一个空列表
lst4=[]
#利用for循环，根据一个列表，创建一个新列表lst4
for i in [1,2,3,1,4,5]:
    lst4.append(i*2+1)
print(lst4)

#删除索引为1的元素
del lst4[1]   #直接从列表中删除，不返回删除的元素
print(lst4)
#删除索引为1的元素，并返回被删除的元素
item=lst4.pop(1)
print(lst4)
print(item)
#pop不指定索引位置，将删除列表最后一个元素
lst4.pop()
print(lst4)
##根据值(或列表元素)来删除
```

```python
lst4= [3, 3, 9]
lst4.remove(3)  #删除第一个匹配的值，大家可思考一下如何删除相同值？
print(lst4)
```

### 3.6 组织列表
对列表排序，Python提供了2个函数：sort、sorted，它们有什么区别呢？  
（1）永久修改列表排序，使用列表函数sort(),使用这种方法不保留底稿；  
（2）使用内置函数sorted()原列表的次序不变，修改后列表存放在一个副本里；  
（3）把列表颠倒过来，使用reverse()函数。  
具体使用请看如下示例：

#### 3.6.1 使用sort()函数

```python
#列表元素为数字##########
lst7=[98,97,87,88,76,68,80,100,65,87,84,92,95,87]
#按数字大小升序排序
print(lst7.sort(reverse=False))
#显示列表
print(lst7)
#列表元素为字符串##########
lst8=["zhangfei","liujing","gaoqiu","wangbo"]
#直接修改列表，返回None值，按字母升序排序
print(lst8.sort(reverse=False))
#显示列表
print(lst8)
```

#### 3.6.2 使用sorted()函数

```python
#列表元素为数字##########
lst7=[98,97,87,88,76,68,80,100,65,87,84,92,95,87]
#按数字大小升序排序
print(sorted(lst7,reverse=False))
#原列表次序不变
print(lst7)
#列表元素为字符串##########
lst8=["zhangfei","liujing","gaoqiu","wangbo"]
#直接修改列表，返回None值，按字母升序排序
print(sorted(lst8,reverse=False))
#原列表次序不变
print(lst8)
```

#### 3.6.3 使用reverse()函数

```python
print(lst8.reverse())
print(lst8)
```

### 3.7 生成列表
前面介绍的列表基本都是手工创建的，用这种方法如果元素不多还可接受，如果要生成几百个、上万个元素就不方便了。这里我们介绍几种生成列表的简单方法，使用这些方法，你可以生成任意多的整数、小数都很方便。

#### 3.7.1range()函数
内置函数range()可以自动生成数据，如果再结合for循环，几乎可以生成任何数据集。range()函数的格式为：
<mark style=background-color:yellow>range (start, stop[, step]) </mark> 

```python
for i in range(5):
    print(i,end=',')
#结果：0,1,2,3,4,
#初始值为1，上限为5
for i in range(1,5):
    print(i,end=',')
#结果：1,2,3,4,
#初始值为2，上限为15，步长为3
for i in range(2,15,3):
    print(i,end=',')
#结果：2,5,8,11,14,
```

#### 3.7.2用range()创建列表

```python
#使用range函数生成列表
lst10=[]
#生成1到9的自然数
for i in range(1,10):
    lst10.append(i)
#清空列表
del lst10[:]    #或lst10.clear()
#生成1到9的平方数
for i in range(1,10):
    lst10.append(i**2)
print(lst10)
```

### 3.8 元组
列表是可修改的序列，其应用比较广泛。有时，我们又希望生成后的列表不能修改，只能读，就像一些只能读的文件一样，用元组就可满足这需求，元组就是不可修改的序列。

#### 3.8.1 定义元组
定义列表用方括号[]，定义元组用圆括号()。定义元组后，就可以使用索引访问元素，这个与列表类似。

```python
#定义一个元组
t1=(2,3,4)
#查看其类型
print(type(t1))
#定义只有一个元素的元组
t2=(10)
#查看t2的类型
print(type(t2))   #显示为int,而不是tuple！
```

<mark style=background-color:yellow>注意：通过t2=(10)生成的t2不是元祖，而是一个数字 ，该如何表示单个元素的元祖呢？</mark> 在元素后加上一个逗号即可，如：

```python
#定义只有一个元素的元组
t2=(10,)
type(t2)    #tuple
#定义空元组
t3=()
type(t3)    #tuple
```

#### 3.8.2 查看元组中元素
查看元组中的元素，与查看列表中元素一样，通过索引就可

```python
#定义一个元组
t3=(2,3,4,5,6)
t3[2] #结果：4
#查看多个元素，注意输出不包括索引4对应的元素，这规则与列表一致
t3[2:4] #结果：4,5
#查看所有元素
t3[:]
```

#### 3.8.3 使用tuple()生成元组
用list()函数可以把range生成的序列转换为列表，与此类似可以用tuple()函数把range生成的序列转换为元组，也可用tuple()函数把列表转换为元组。

```python
a1=tuple(range(5))
#查看a1的类型
print(type(a1))
lst=["华为","苹果","三星"]
t1=tuple(lst)
```

#### 3.8.4 元组特殊用途
元组可作为字典中键来使用(字典那章将介绍，列表不能作为字典的键)，内建函数、自定义函数的返回值（如enumerate(a1)）大多都是元组形式。

#### 3.8.5  元组解包

```python
a,b,c=(7,9,100)
print(a)  #7
print(b)  #9
print(c)  #100
```

在深度学习中经常遇到处理图像数据与标签放在一个元组中的问题，即数据和标签放在一个元组中，然后，通过元组解包的方式，把数据、标签分别存放在两个变量中，如下例：

```python
#构建一个列表，其元素由元组构成
#每个元组由表示图像数据的列表及一个表示图像标签的数据构成
lst31=[([0.11,0.26],'0'),([1.21,0.78],'1'),([1.31,0.48],'0'),([0.25,0.37],'1')]
#对每个元组进行解包，得到图像数据和标签
for image_data,label in lst31:   
    print("图像数据:{},标签:{}".format(image_data,label))
```
